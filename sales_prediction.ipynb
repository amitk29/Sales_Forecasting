```python
from fbprophet import Prophet
```

```python
def sales_predictions(Store_ID,market_df,periods):
    market_df=market_df[market_df['Store']==Store_ID]
    market_df=market_df[['Date','Sales']].rename(columns={'Date':'ds','Sales':'y'})
    market_df=market_df.sort_values('ds')
    
    model=Prophet()
    model.fit(market_df)
    future=model.make_future_dataframe(periods=periods)
    
    forecast=model.predict(future)
    figure=model.plot(forecast, xlabel='Date', ylabel='Sales')
    
    figure2=model.plot_components(forecast)
    
    
```

```python
sales_predictions(10,market_df,60)
```
